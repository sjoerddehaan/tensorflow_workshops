## Hardware support
- CPU
- GPU
- TFU
- Clusters
- Mobile

TensorFlow runs on ordinary computers. However, it can be slow for applications like deep learning, which involve lots of matrix manipulations. GPU's are better suited for this sort of work. GPU stands for 'graphical processing unit', as they were originally designed for manipulating computer graphics. GPU's can rapidly manipulate and alter memory to accelerate the creation of images in a frame buffer. The parallel design makes them them efficient for processing large blocks of data in parallel; exactly what we do in machine learning.

The TPU is Googles propriety hardware for deep learning. TPU's are designed for high volumes of low precision computations. Google prefers it's own design, as it is both fast and energy efficient.

Training of machine learning algorithms can be computationally intensive, even on a GPU or TPU. Luckily, many learning algorithms can be parallelised to speed up the learning process. TensorFlow can be ran in a distributed manner, for example in a data center in the cloud.

At last, TensorFlow runs on mobile devices, enabling developers to make use of TensorFlow models in mobile apps in production.

> The tensor is the generalization of vectors and matrices

A vector is a tensor of rank 1  
A matrix is a tensor or rank 2

## Computation graphs in TensorFlow


> A computation graph **defines a calculation**

- It does not perform the calculation
- It is a blue print for one or more computations
- It lays out the flow of data

## Advantages of the computation graph




A computation graph is like the recipe of your favorite cake 

- You can bake it whenever you like
- You can bake it in whatever kitchen you have available
- You can bake cakes as small or as big as you wish
- You can experiment with different proportions, temperature, baking time, etc
- You can decide to omit the icing or leave out the berry sauce
- You can use different ingredients; e.g. apple instead of pear
- You can bake multiple cakes at the same time
- You can even start your own cake factory


A computation graph is a very usefull abstraction when you are doing machine learning.

- You can push calculations to whatever hardware you have available
- You can perform calculations from any location by making use of the cloud
- You can repeat your computation as many times as you need (useful for optimization problems)
- You can experiment with varying the parameters of your algorithm
- You can execute or repeat a selected part of the computation graph instead of running the whole graph
- You can re-do your calculation with different inputs; e.g. a different batches of users
- You can use multiple machines at the same time to speed up your job
- You push your newly developed model into production easily; e.g use it in your mobile app

## Sum of squares - example of `Variable` 
Maintain a running sum of squares.

```python
import tensorflow as tf
x = tf.placeholder(dtype=tf.float32)
square = tf.pow(x, 2)
sum_of_squares = tf.Variable(0, dtype=tf.float32) # Create variable node
update_sum = tf.assign(sum_of_squares, sum_of_squares + square) # Create update operator
with tf.Session() as sess:
    sum_of_squares.initializer.run() # Assign initial value (0) to variable
    print(sum_of_squares.eval())
    print(sess.run([square, sum_of_squares], feed_dict={x: 1}))
    print(sess.run([square, update_sum, sum_of_squares], feed_dict={x: 2}))
    print(sess.run([update_sum], feed_dict={x: 3}))
    print(sum_of_squares.eval())
    print(update_sum.eval(feed_dict={x: 2}))
    print(sum_of_squares.eval())
```

- Variables must be initialized explicitly before a graph is used for the first time
- A common use of variables is to modify them after each iteration, as we search for the models optimal parameter settings

````python
import tensorflow as tf
x = tf.placeholder(dtype=tf.float32)
square = tf.pow(x, 2)
sum_of_squares = tf.Variable(0, dtype=tf.float32) # Create variable node
update_sum = tf.assign(sum_of_squares, sum_of_squares + square) # Create update operator
with tf.Session() as sess:
    sum_of_squares.initializer.run() # Assign initial value (0) to variable
    print(sum_of_squares.eval())
    print(sess.run([square, sum_of_squares], feed_dict={x: 1}))
    print(sess.run([square, update_sum, sum_of_squares], feed_dict={x: 2}))
    print(sess.run([update_sum], feed_dict={x: 3}))
    print(sum_of_squares.eval())
    print(update_sum.eval(feed_dict={x: 2}))
    print(sum_of_squares.eval())
````

# Case study: moving averages with TensorFlow

## Simple Moving average

Keep an average over a sequence of numbers  
This results in a new  sequence  

Example  
Take the sequence 1, 2, 2, 1  
What is the sequence of moving averages? 

Answer  
Take the average of all number so far  
1, (1+2)/2, (1+2+2)/3, (1+2+2+1)/4

## Exponential moving average
This time we use weights that are not equal  
We give higher weights to the most recent observations

````python
from IPython.display import IFrame
HTML('<p><a href="https://commons.wikimedia.org/wiki/File:Exponential_moving_average_weights_N%3D15.png#/media/File:Exponential_moving_average_weights_N%3D15.png"><img width=400 src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/ba/Exponential_moving_average_weights_N%3D15.png/1200px-Exponential_moving_average_weights_N%3D15.png" alt="Exponential moving average weights N=15.png"></a><br><a href="http://creativecommons.org/licenses/by-sa/3.0/" title="Creative Commons Attribution-Share Alike 3.0">CC BY-SA 3.0</a>, <a href="https://commons.wikimedia.org/w/index.php?curid=582114">Link</a></p>')
````

## Formula for exponential moving average 

The average $S_t$ of the original sequence $Y_t$ is given by
$$
{\displaystyle S_{t}={\begin{cases}Y_{1},&t=1\\\alpha \cdot Y_{t}+(1-\alpha )\cdot S_{t-1},&t>1\end{cases}}}
$$
See [wikipedia](https://en.wikipedia.org/wiki/Moving_average#Exponential_moving_average)

## Tensorflow example

```Python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

# Python variables 
l = 100 # Number of data points
raw_data = np.random.normal(10, 1, l) # Create sequence of random values
moving_average = np.zeros(raw_data.shape) # Initial sequence of moving averages
alpha_val = 0.1 # Decay factor
```

```Python
# Tensorflow nodes (variables, constants, place holders, and operators)
alpha = tf.constant(alpha_val, name='alpha')
y_t = tf.placeholder(tf.float32) # Used to insert latest raw value into graph
s_tminus = tf.Variable(0.) # previous value of moving average (s_{t-1})
s_t = alpha * y_t + (1 - alpha) * s_tminus # Operator for updating 
update = tf.assign(s_tminus, s_t)
init = tf.global_variables_initializer() # Operator for setting initial values
```

```Python
# Perform the calculation of the moving average within the Tensorflow session
with tf.Session() as sess:
    sess.run(init)
    for i in range(len(raw_data)): # Python loop 
        s_t_value = sess.run(update, feed_dict={y_t:raw_data[i]})
        moving_average[i] = s_t_value # Store latest value of s_t
```

````python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

# Python variables 
l = 100 # Number of data points
raw_data = np.random.normal(10, 1, l) # Create sequence of random values
moving_average = np.zeros(raw_data.shape) # Initial sequence of moving averages
alpha_val = 0.1 # Decay factor

# Tensorflow nodes (variables, constants, place holders, and operators)
alpha = tf.constant(alpha_val, name='alpha')
y_t = tf.placeholder(tf.float32) # Used to insert latest raw value into graph
s_tminus = tf.Variable(0.) # previous value of moving average (s_{t-1})
s_t = alpha * y_t + (1 - alpha) * s_tminus # Operator for updating 
update = tf.assign(s_tminus, s_t)
init = tf.global_variables_initializer() # Operator for setting initial values


# Perform the calculation of the moving average within the Tensorflow session
with tf.Session() as sess:
    sess.run(init)
    for i in range(len(raw_data)): # Python loop 
        s_t_value = sess.run(update, feed_dict={y_t:raw_data[i]})
        moving_average[i] = s_t_value # Store latest value of s_t

# Plot the data     
plt.plot(moving_average, label=r"EWMA $\alpha$ = {}".format(alpha_val))
plt.plot(raw_data, label="Raw data")
plt.title(r"Moving averages".format(alpha_val))
plt.legend();
````

````python
import tensorflow as tf
import numpy as np

raw_data = np.random.normal(10,1, 100)
alpha = tf.constant(0.05)

# Nodes for the computation graph
curr_value = tf.placeholder(tf.float32)
prev_avg = tf.Variable(0.)
update_avg = alpha * curr_value + (1-alpha) * prev_avg
avg_hist = tf.summary.scalar("running_average", update_avg) # Create a summary operation with tag running_average
value_hist = tf.summary.scalar("incoming_value", curr_value) # Create a summary operation with tag incoming_values
merged_summary = tf.summary.merge_all() # Combine all summary ops into a single op that we will run later
init = tf.global_variables_initializer() # Create initialization operator (operator is created only, not executed)

# Setup summary writer
graph = tf.get_default_graph() # Optional for storing the graph
writer = tf.summary.FileWriter("log/moving_average", graph=graph)

with tf.Session() as sess: # Create a new tensorflow session
    sess.run(init) # Execute initialization operator in this session
    for i in range(len(raw_data)):
        summary_str, curr_avg = sess.run([merged_summary, update_avg], feed_dict=\
                                         {curr_value: raw_data[i]}) # Execute operators merged_summary and update_avg
        sess.run(tf.assign(prev_avg, curr_avg)) # Execute operator 
        writer.add_summary(summary_str, i) # Write the summary string, together with the gobal step i
        
````
